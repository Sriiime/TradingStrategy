```python
# Set strategy parameters
short_window = 50
long_window = 200

# Calculate moving averages
df['Short_MA'] = df['Close'].rolling(window=short_window).mean()
df['Long_MA'] = df['Close'].rolling(window=long_window).mean()

# Create signals
df['Signal'] = np.where(df['Short_MA'] > df['Long_MA'], 1, 0)

# Calculate strategy returns
df['Strategy_Returns'] = df['Close'].pct_change() * df['Signal'].shift(1)

# Evaluate strategy performance
sharpe_ratio = df['Strategy_Returns'].mean() / df['Strategy_Returns'].std() * np.sqrt(252)
sortino_ratio = df['Strategy_Returns'].mean() / df['Strategy_Returns'][df['Strategy_Returns'] < 0].std() * np.sqrt(252)
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Sortino Ratio: {sortino_ratio:.2f}")

# Plot strategy performance
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['Close'] * df['Signal'], label='Strategy')
plt.legend()
plt.show()

# Refine strategy
def optimize_strategy_params(df, short_window_range, long_window_range):
    def objective(params):
        short_window, long_window = np.int_(params)
        df['Short_MA'] = df['Close'].rolling(window=short_window).mean()
        df['Long_MA'] = df['Close'].rolling(window=long_window).mean()
        df['Signal'] = np.where(df['Short_MA'] > df['Long_MA'], 1, 0)
        df['Strategy_Returns'] = df['Close'].pct_change() * df['Signal'].shift(1)
        return -df['Strategy_Returns'].mean() / df['Strategy_Returns'].std()

    bounds = [(int(short_window_range[0]), int(short_window_range[1])),
              (int(long_window_range[0]), int(long_window_range[1]))]
    res = minimize(objective, [short_window, long_window], bounds=bounds)
    return res.x

optimal_params = optimize_strategy_params(df, (20, 100), (100, 300))
print(f"Optimal Parameters: {optimal_params}")
```

```python
# Define strategy parameters
short_window = 50
long_window = 200

# Define parameter ranges
short_window_range = (20, 100)
long_window_range = (100, 300)

# Define objective function
def objective(params):
    short_window, long_window = params
    df['Short_MA'] = df['Close'].rolling(window=short_window).mean()
    df['Long_MA'] = df['Close'].rolling(window=long_window).mean()
    df['Signal'] = np.where(df['Short_MA'] > df['Long_MA'], 1, 0)
    df['Strategy_Returns'] = df['Close'].pct_change() * df['Signal'].shift(1)
    return -df['Strategy_Returns'].mean() / df['Strategy_Returns'].std()

# Run parameter stability testing
results = []
for short_window in range(int(short_window_range[0]), int(short_window_range[1])):
    for long_window in range(int(long_window_range[0]), int(long_window_range[1])):
        params = [short_window, long_window]
        result = objective(params)
        results.append((short_window, long_window, result))

# Plot results
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.scatter([x[0] for x in results], [x[1] for x in results], c=[x[2] for x in results])
plt.title('Parameter Stability Testing')
plt.xlabel('Short Window')
plt.ylabel('Long Window')
plt.colorbar(label='Strategy Performance')
plt.show()
```

```python
# Define strategy parameters
short_window = 50
long_window = 200

# Define Monte Carlo simulation parameters
n_simulations = 1000
n_days = len(df)

# Initialize results array
mc_results = np.zeros((n_simulations, n_days))

# Run Monte Carlo simulations
for i in range(n_simulations):
    # Generate random market conditions
    market_returns = np.random.normal(0, 0.05, size=n_days)

    # Calculate strategy returns
    strategy_returns = np.where(df['Short_MA'] > df['Long_MA'], market_returns, 0)

    # Store results
    mc_results[i] = strategy_returns

# Calculate average strategy returns across simulations
avg_strategy_returns = np.mean(mc_results, axis=0)

# Plot results
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(avg_strategy_returns)
plt.title('Monte Carlo Simulation Results')
plt.xlabel('Days')
plt.ylabel('Strategy Returns')
plt.show()
```
